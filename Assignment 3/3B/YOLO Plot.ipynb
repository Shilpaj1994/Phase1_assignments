```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

x = np.arange(0, 22, 1)
y = [32, 64, 128, 64, 128, 256, 128, 256, 512, 256, 512, 256, 512, 1024, 512, 1024, 512, 1024, 1024, 1024, 64, 1024]
maxpool_layer = [1, 2, 5, 8, 13]

# Plot the points using matplotlib
plt.figure(figsize=(15,4))
plt.xlabel('Layer Number')
plt.ylabel('Number of Kernels')
plt.title('YOLO classification network')
plt.xticks(x)
plt.plot(x, y, '-o', markevery=maxpool_layer)
plt.legend(['Max_pooling layer'])
plt.savefig('YOLO')
plt.show()
```
